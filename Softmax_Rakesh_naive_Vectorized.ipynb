```python
import numpy as np
```

```python
X=np.array([[240,22,153,100],[35,213,45,120],[100,45,67,25]])
print('images\n',X)
W=np.array([[.2,.5,.9],[.1,.5,.8],[.3,.7,.4],[.6,.9,.1]])
print('weights\n',W)
y=(1,0,2)
print('labels\n',y)
```

# Softmax naive

```python
dW=np.zeros(W.shape)
num_train = X.shape[0]
print('num of training images\n',num_train)
num_classes = W.shape[1]
print('num of classes\n',num_classes)
loss = 0.0
for i in range(num_train):
    print('i=',i)
    scores = X[i].dot(W)
    print('scores\n',scores)
    scores -= np.max(scores)
    print('Scores\n',scores)
    correct_class_score= scores[y[i]]
    print('correct_class_score\n',correct_class_score)
   
    sum_j = np.sum(np.exp(scores))
    print('sum_j\n',sum_j)
    softmax =np.exp(correct_class_score) / sum_j
    print('softmax\n',softmax)
    loss += -np.log(softmax)
    print('loss\n',loss)
    
    for j in range(num_classes):
        print('j == y[',i,']')
        if j == y[i]:
            print('scores[j]\n',scores[j])
            
            dW[:, y[i]] += ((np.exp(scores[j]) / sum_j) - 1) * X[i, :]
            print('dW after X[',i,']=',X[i],' is multiplied with softmax correct class ',j,' coordinate\n',dW)
        else:
            print('scores[j]\n',scores[j])
            dW[:, j] += ((np.exp(scores[j]))/ sum_j) * X[i, :]
            print('dW after X[',i,']=',X[i],' is multiplied with softmax other class',j,' coordinate\n',dW)
    
loss /= num_train
#print('loss\n',loss)
dW /= num_train
#print('dW \n',dW )

```

# Softmax vectorized

```python
loss = 0.0
dW = np.zeros_like(W)
print('dW\n',dW)
   
num_train = X.shape[0]
scores = X.dot(W)
print('scores\n',scores)
print('dimscores\n',scores.shape)
scores = scores - np.max(scores, axis=1)[:,np.newaxis]

print('scores\n',scores)
print('dimscores\n',scores.shape)
softmax= (np.exp(scores))/ np.sum(np.exp(scores), axis=1)
print('softmax\n',softmax)

loss += -np.sum(np.log(softmax[np.arange(num_train), y])) 
print('loss\n',loss)
loss /= num_train

intg = np.zeros_like(scores)
print('intg\n',intg)
intg[np.arange(num_train), y] = 1
print('intg\n',intg)
dW = X.T.dot(softmax-intg)
print('dW\n',dW)

loss /= num_train
#print('loss\n',loss)
dW /= num_train
#print('dW \n',dW )
```
